```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def import_file(tax):
    if ".csv" in tax:
        taxa_sum = pd.read_csv(tax, sep=",", header=0, index_col=0)
    elif '.tab' or '.tsv' in tax:
        taxa_sum = pd.read_csv(tax, sep="\t", header=0, index_col=0)
    else: 
        print('unsupported file type')
    return taxa_sum
```

```python
file = import_file('tax.summary.all.tab')
```

```python
file.groupby(file.index).mean()
```
